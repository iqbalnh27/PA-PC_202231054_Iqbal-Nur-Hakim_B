```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk menampilkan gambar menggunakan Matplotlib
def show_image(title, image, cmap_type='gray'):
    plt.figure(figsize=(5, 5))
    plt.title(title)
    plt.imshow(image, cmap=cmap_type)
    plt.axis('off')
    plt.show()
```

```python
# Membaca gambar asli
image_path = 'daunbaru.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
show_image('Gambar Asli', image_rgb, 'viridis')
```

```python
# Mengonversi gambar ke dalam format HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
# Membuat masker untuk mangga (kuning)
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])
mask_mango = cv2.inRange(image_hsv, lower_yellow, upper_yellow)
show_image('Mask Mangga', mask_mango, 'gray')
```

```python
# Membuat masker untuk daun (hijau)
lower_green = np.array([35, 100, 100])
upper_green = np.array([85, 255, 255])
mask_leaf = cv2.inRange(image_hsv, lower_green, upper_green)
show_image('Mask Daun', mask_leaf, 'gray')
```

```python
# Segmentasi mangga
segmented_mango = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_mango)
show_image('Segmentasi Mangga', segmented_mango, 'viridis')
```

```python
# Segmentasi daun
segmented_leaf = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_leaf)
show_image('Segmentasi Daun', segmented_leaf, 'viridis')
```
